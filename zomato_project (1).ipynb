#**ZOMATO DATA ANALYSIS PROJECT**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Zomato data .csv')
print(data)
```

```python
data
```

#Convert the data type of column-rate

```python
def handleRate(value):
  value= str(value).split('/')
  value= value[0];
  return float(value)

data['rate'] = data['rate'].apply(handleRate)
print(data.head())
```

```python
data.info()
```

#Type of Restaurant

```python
data.head()
```

```python
# Calculate counts
counts = data['listed_in(type)'].value_counts()

# Plot using Matplotlib
plt.figure(figsize=(6,6))
plt.bar(counts.index, counts.values, color='blue')
plt.xlabel("Type of Restaurant")
plt.ylabel("Number of Listings")
plt.title("Count of Each Type of Restaurant")
plt.xticks(rotation=45)
plt.show()
```

Conclusion:Majority of the restaurant falls in  dinning category

#No. of Votes For Restaurant

```python
data.head()
```

```python
grouped_data=data.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green",marker="o")
plt.xlabel("Type of restaurant",c="blue",size=20)
plt.ylabel("Votes",c="blue",size=20)
```

Conclusion:Dinning Restaurant has received maximum votes.

#Ratings of majority of Restaurant

```python
data.head()
```

```python
plt.hist(data['rate'],bins=10)
plt.title("Ratings Distribution")
plt.show()
```

Conclusion:The majority Restaurant receive ratings from 3.5 to 4.0

#Average Spending on Food by Couples

```python
data.head()
```

```python
# Calculate counts
cost_counts = data['approx_cost(for two people)'].value_counts()

# Plot using Matplotlib
plt.figure(figsize=(8, 5))  # Adjust size as needed
plt.bar(cost_counts.index.astype(str), cost_counts.values, color='green')
plt.xlabel("Approximate Cost for Two People")
plt.ylabel("Number of Listings")
plt.title("Count of Each Approximate Cost for Two People")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Conclusion:Majority of couples prefer Restaurant with an approximate cost of 300 rs

#Which mode receive maximum Rating

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Prepare the data
grouped_data = data.groupby('online_order')['rate'].apply(list)

# Create a figure with specific size
plt.figure(figsize=(6, 6))

# Create the boxplot
plt.boxplot([grouped_data['Yes'], grouped_data['No']],
            labels=['Yes', 'No'])

# Add labels and title
plt.xlabel("Online Order")
plt.ylabel("Rate")
plt.title(" Ratings by Online Order")

# Show the plot
plt.show()
```

Conclusion:Offline orders receive lower Ratings in comparison to online order.

#Heatplot

```python
data.head()
```

```python
pivot_table=data.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
plt.imshow(pivot_table, cmap="YlGnBu", aspect='auto')

# Add color bar
plt.colorbar(label='Count')

# Add annotations (optional)
for i in range(pivot_table.shape[0]):
    for j in range(pivot_table.shape[1]):
        plt.text(j, i, pivot_table.iat[i, j], ha='center', va='center', color='black')
# Add labels and title
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.title("Heatmap")

# Set x and y ticks
plt.xticks(ticks=np.arange(pivot_table.columns.size), labels=pivot_table.columns)
plt.yticks(ticks=np.arange(pivot_table.index.size), labels=pivot_table.index)

# Display the plot
plt.show()
```

Conclusion:Dinning Restaurant primarily accept offline orders,whereas cafes primarily receive online orders.This suggest that client prefer order in person at restaurants,but prefer online ordering at cafes.
